```python
AirBnB_df = pd.read_csv('Resources/AB_US_2020.csv')
AirBnB_df
```

```python
#Clean Data
last_review = AirBnB_df.loc[:,'last_review']
AirBnB_df['last_review'] = AirBnB_df['last_review'].fillna('00/00/00')
AirBnB_df["last_review"].fillna('00/00/00', inplace = True)

AirBnB_df.head()
```

```python
i=0
for date in last_review:
    if int(date[:2]) > 12:
        split_date = date.split('/')
        new_date = split_date[1] + '/' + split_date[0] + '/' + split_date[2]
        AirBnB_df.loc[i,'last_review'] = new_date
    else:
        next
    i = i + 1

AirBnB_df.head()
```
